## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

The starter dataset contains 30,000 pages, and 16,262 of them are marked as declining, which is 54.2% of the dataset. The median CTR is 0.070, or about 7%. These observed numbers suggest that declining performance is common enough to justify building a prioritization system. Because more than half of the pages are declining, the system should not simply flag every declining page; it should help rank which pages deserve attention first.

```python
import pandas as pd

data_url = (
    "https://raw.githubusercontent.com/"
    "ahsan-rajput/flyrank-ml-internship-portfolio/"
    "main/data/raw/content_refresh_anonymized.csv"
)

df = pd.read_csv(data_url)

total_pages = len(df)
declining_pages = (df["trend_direction"] == "down").sum()
declining_percent = declining_pages / total_pages * 100
median_ctr = df["ctr"].median()

print(f"Total pages in the starter data: {total_pages:,}")
print(f"Pages marked as declining: {declining_pages:,}")
print(f"Percentage declining: {declining_percent:.1f}%")
print(f"Median CTR: {median_ctr:.3f}")
```
